## Analyzie resident file & uncompressed_len


```python
import pandas as pd
import numpy as np
```

```python
filepath = "/content/drive/MyDrive/security_mft/output/1123202101/parse_result_temp_byte_runs_1123202101.csv"
df = pd.read_csv(filepath)
df.head()
```

```python
grouped = df.groupby('type')
grouped.describe()
```

```python
df_resident = df.loc[df['type'] == 'resident']
df_resident.head()
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df_resident_notna = df_resident[df_resident['uncompressed_len'].notna()]
print(df_resident_notna.head())
print(df_resident_notna.shape[0]) # number of rows
```

# Sqlite


```python
import pandas as pd
import sqlite3 as sq
data = df_resident_notna
sql_data = '/content/drive/MyDrive/security_mft/output/1123202101/resident_notna.sqlite' #- Creates DB names SQLite
conn = sq.connect(sql_data)
cur = conn.cursor()
cur.execute('''DROP TABLE IF EXISTS resident_notna''')
data.to_sql('resident_notna', conn, if_exists='replace', index=False) # - writes the pd.df to SQLIte DB
df_read_sql = pd.read_sql('select * from resident_notna', conn)
conn.commit()
conn.close()
```
