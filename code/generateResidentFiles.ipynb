```python
import pandas as pd
import numpy as np
import os
import glob
import sys
import hashlib
# from random import randbytes #python == 39
# from natsort import natsorted # pip install natsort # used for sort str+num
# import math
```

```python
pathToResidentFiles = "/content/drive/MyDrive/security_mft/resident_files"
```

```python

df = pd.DataFrame(columns=['filename','hashvalue','size'])
```

```python
def generateFiles(numberOfFiles, sizeBytes):
  for i in range(numberOfFiles):
    filename = f"{sizeBytes}_bytes_{i}"
    strContent = os.urandom(sizeBytes) 
    with open(f"{pathToResidentFiles}/{filename}", "wb") as f:
      f.write(strContent)

    
    with open(f"{pathToResidentFiles}/{filename}", "rb") as f:
      md5_hash = hashlib.md5()
      content = f.read()
      md5_hash.update(content)
      digest = md5_hash.hexdigest()
      print(f"hash value: {digest}")
      global df
      df = df.append({'filename': filename, 
                          'hashvalue':digest, 
                          'size':sizeBytes
                          }, 
                         ignore_index=True)
    
```

```python
generateFiles(2, 100)
generateFiles(1, 200)
generateFiles(4, 500)

for eachFile in glob.glob(f"{pathToResidentFiles}/*"):
  print(f"size (bytes): {os.path.getsize(eachFile)}")
```
